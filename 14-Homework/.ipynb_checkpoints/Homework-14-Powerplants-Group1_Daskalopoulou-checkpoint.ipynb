## Make a map of power plants in the US, color coded by type

```python
points = df_pp.apply(lambda row: Point(row.Latitude, row.Longitude), axis=1)
geo_pp = gpd.GeoDataFrame(df_pp, geometry=points)
geo_pp.crs = {'init':'epsg:5071'}
```

```python
geo_pp.shape
```

```python
ax = states.plot(color='lightgrey', linewidth=0.25, edgecolor='white')
geo_pp.plot(ax=ax, color='red', markersize=3)
ax.axis('off')
```

```python
# if the code above worked I would also do 
ax = states.plot(color='lightgrey', linewidth=0.25, edgecolor='white')
geo_pp.plot(column='PrimSource', cmap='OrRd', markersize=3, legend=True, ax=ax)
ax.axis('off')
```

## Make a map of air quality stations, color coded by worst air quality ever 

```python
points = df_stations.apply(lambda row: Point(row.Latitude, row.Longitude), axis=1)
geo_stations = gpd.GeoDataFrame(df_stations, geometry=points)
geo_stations.crs = {'init':'epsg:5071'}
```

```python
ax = states.plot(color='lightgrey', linewidth=0.25, edgecolor='white')
geo_stations.plot(column='1st Max Value', cmap='OrRd', markersize=3, legend=True, ax=ax)
ax.axis('off')
```

## Map all of the nuclear power plants in red, with the rest of the plants as dark grey.


```python
st_nuc = geo_pp[(geo_pp.PrimSource == 'nuclear')]
st_else = geo_pp[(geo_pp.PrimSource != 'nuclear')]
```

```python
ax = states.plot(color='lightgrey', linewidth=0.25, edgecolor='white')
st_nuc.plot(column='PrimSource', color='red', markersize=3, legend=True, ax=ax)
st_else.plot(column='PrimSource', color='darkgrey', markersize=3, legend=True, ax=ax)
ax.axis('off')
```
