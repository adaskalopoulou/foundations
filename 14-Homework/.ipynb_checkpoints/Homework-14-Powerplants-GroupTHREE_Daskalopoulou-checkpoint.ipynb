## 3. Map all of the power plants in New York, New Jersey, Pennsylvania and Connecticut. Have a nice bottom layer of those 4 states.

```python
my_states = ['New York','New Jersey','Pennsylvania','Connecticut']
```

```python
east_states = states[states.NAME.isin(my_states)]
```

```python
ax = east_states.plot(color='grey', linewidth=0.25, edgecolor='white', figsize=(20, 20))
geo_pp[geo_pp.intersects(east_states.unary_union)].plot(ax=ax, color='blue', markersize=5)
ax.axis('off')
```
