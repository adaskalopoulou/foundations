## Oh, weird, ABV isn't a number. Convert it to a number for me, please.

It's going to take three steps:

1. List just the ABV of every row
2. Remove the % symbol
3. Convert it to a `float`
4. Save it into a new column. Or the same column, if you're feeling wild.

# Measuring International Bitterness Units

IBU stands for [International Bitterness Unit](http://www.thebrewenthusiast.com/ibus/), and while a lot of places like to brag about having the most bitter beer (it's an American thing!), IBUs don't necessary *mean anything*.

Let's look at how different beers have different IBU measurements.

## What is the average IBU measurement across all beers?

```python
df["IBUs"].mean().astype(int)
```

## Draw the distribution of IBU measurements, but with *thirty* bins instead of the default of 10

```python
%matplotlib inline
```

```python
df["IBUs"].hist(bins=30)
```

## List all of the beers with IBUs above the 75th percentile

```python
df["IBUs"].describe(percentiles=None, include=None, exclude=None)
```

```python
df[df.IBUs > 64]
```

## List all of the beers with IBUs below the 25th percentile

```python
df[df.IBUs < 21]
```

## List the median IBUs of each type of beer. Graph it.

Put the highest at the top, and the nan ones at the bottom. I want a nice graph, too - don't let them all be squished together, either!

```python
df.groupby("Style")["IBUs"].mean()
```

```python
df.groupby("Style")["IBUs"].mean().dropna().sort_values().tail(20).plot(kind='barh')
```

To look nicer I graphed only the 20 most bitter beers.

```python
df.groupby("Style")["IBUs"].mean().dropna().sort_values().head(20).plot(kind='barh')
```

Also, here I graphed the 20 less bitter beers. 

## Hmmmm, it looks like they are generally different styles. What are the most common 5 styles of high-IBU beer vs. low-IBU beer?

You'll want to think about it in three pieces - finding the high IBU beers, then finding out what the most common styles are, then getting the top 5.

```python
df.groupby("Style")["IBUs"].mean().dropna().sort_values().tail(5)
```

```python
df.groupby("Style")["IBUs"].mean().dropna().sort_values().head(5)
```

## Get the average IBU of "Witbier", "Hefeweizen" and "American Pale Wheat Ale" styles (also draw a histogram)

I want ONE measurement and ONE graph, not three separate ones. And 20 to 30 bins in the histogram, please.

```python
choice_beer = df[(df['Style'] == "Witbier") | (df['Style'] == "Hefeweizen") | (df['Style'] == "American Pale Wheat Ale")]
```

```python
choice_beer["IBUs"].mean().astype(int)
```

```python
choice_beer["IBUs"].plot(kind='hist', bins=30)
```

## Get the average IBU of any style with "IPA" in it (also draw a histogram)

```python
IPA_beer = df[df['Style'].str.contains("IPA", na=False)]
```

```python
IPA_beer["IBUs"].mean().astype(int)
```

```python
IPA_beer["IBUs"].plot(kind='hist', bins=30)
```

## Plot those two histograms on top of one another

```python
ax = choice_beer["IBUs"].plot(kind='hist', bins=30)
IPA_beer["IBUs"].plot(kind='hist', bins=30, ax=ax)
```

## Compare the ABV of wheat beers vs. IPAs 

```python
IPA_beer['ABV_clean'].mean()
```

```python
wheat_beer = df[df['Style'].str.contains("Wheat", na=False)]
wheat_beer["ABV_clean"].mean()
```

The ABV of wheat beers is lower than the ABV of IPAs.
