```python
import pandas as pd
```

```python
df= pd.read_csv("tabula-ACFPPricingDoc2016-6-8.csv", encoding='latin1')
df
```

```python
bad_df = df[df['Dosing'].isnull()]
bad_df
```

```python
df.loc[df['Dosing'].isnull(), 'Type'] =  df['Generic Name']
df
```

```python
df['Type'].fillna(method='ffill', inplace = True)
df.head()
```

```python
new_df = df.dropna(subset=['Dosing'])
new_df
```

```python
new_df['90 Day Cost'] = new_df['90 Day Cost'].str.replace("$","")
```

```python
new_df['90 Day Cost'] = new_df['90 Day Cost'].astype(float, inplace=True)
```

```python
new_df['90 Day Cost'].dtypes
```

## What is the most expensive type of cardiovascular drugs?

```python
new_df.groupby('Type')['90 Day Cost'].mean().sort_values(ascending=False)
```
